```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
mpl.rcParams['font.family'] = ['Roboto']
```

```python
df_time = pd.read_csv('result/positivity_bytime.csv')
```

```python
df_time.head()
```

```python
fig, ax = plt.subplots(figsize=(13,5))

ax2 = ax.twinx()
ax.plot(df_time['lab_date'], df_time['pos_rate'], 'g.-', color='red')
ax2.bar(df_time['lab_date'], df_time['total_num'], alpha=0.2, color='skyblue')

# for tick in ax.xaxis.get_minor_ticks():
#     tick.lab_date.set_xticklabels('right')

ax.set_xticklabels(df_time['lab_date'], rotation=90, fontsize=10);#, rotation_mode='anchor'
ax.set_ylabel('Positivity Rate', fontsize=18-0);
ax2.set_ylabel('Number of Tests', rotation=270, fontsize=18-0);
ax.set_xlim(-1,40);
ax2.yaxis.set_label_coords(1.07,0.5);
ax.set_title('Positivity Rate over time', fontsize=20)
ax.tick_params(axis='y', colors='red')
ax2.tick_params(axis='y', colors='skyblue')
ax.yaxis.label.set_color('red')
ax2.yaxis.label.set_color('skyblue')

# for tick in ax.xaxis.get_majorticklabels():
#     tick.set_y(-.08)

plt.savefig('figs/posrate.png', bbox_inches='tight')
```
